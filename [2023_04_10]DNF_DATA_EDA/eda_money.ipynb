## 데이터 이상 탐지
### 1. RLS 필터를 이용한 데이터 이상 탐지
- RLS 필터 클래스

```python
class RLSFilter:
    def __init__(self, windows, first_value, lambda_=0.99, delta = 1e-6):
        self.windows = windows # 필터 크기
        self.lambda_ = lambda_ # 이 값이 커질수록 이전 데이터의 영향력이 크고, 작아질수록 현재 데이터의 영향력이 커집니다. forgetting factor. 0< lmbda <= 1. 일반적으로 0.98에서 1 사이의 값으로 선택함.
        self.delta = delta # 초기 공분산 행렬 P의 대각선 요소들이 0이 되지 않게 하는 값. 0이 되면 역행렬을 구할수 없으므로.
        self.buffer = np.ones(windows) * first_value# 입력값을 저장하기 위한 변수
        self.P = np.eye(windows) * (1 / delta) # 칼만 게인 계산하기 위한 행렬
        self.theta = np.zeros(windows) # 초기 가중치 값
    # 데이터 업데이트
    def update(self, x, y):
        # left pop 후 append. queue 형식.
        self.buffer[:-1] = self.buffer[1:]
        # 메모리 참조 이슈를 피하기 위해 copy 사용.
        # self.buffer[:-1] = self.buffer[1:].copy()
        self.buffer[-1] = x
        P = self.P
        theta = self.theta

        # 칼만 게인 계산
        temp1 = self.P @ self.buffer
        temp2 = (self.lambda_ + (self.buffer @ self.P) @ self.buffer)
        K = temp1 / temp2

        # 추정치에 곱할 가중치 업데이트
        e = y - self.theta @ self.buffer
        self.theta += K * e

        # 공분산 행렬 업데이트. Sherman–Morrison formula 사용.
        self.P = (self.P - K.reshape(-1, 1) @ self.buffer.reshape(1, -1) @ self.P) / self.lambda_
        
        # 데이터 예측 값 반환
        return self.theta @ self.buffer
```

- 이상탐지를 위한 던파 네이버 카페 거래 내역 데이터 load

```python
spark = SparkSession.builder \
    .appName("DNFData") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()

df_data = spark.read.format("csv").option("header", "true").load("dnf_cafe_money.csv")
df_data.createOrReplaceTempView("user_money")
result = spark.sql("SELECT * FROM user_money")
result.show()
```

- spark 데이터 프레임에서 pandas 데이터 프레임으로 변환 후 데이터 형변환

```python
import pandas as pd
# date와 value의 값을 따로 처리.
date = spark.sql("SELECT date FROM user_money")
value = spark.sql("SELECT value FROM user_money")
# 데이트 타입 변환
date = date.toPandas().astype({'date':'int64'}).values
date = pd.to_datetime(np.ravel(date), format='%Y%m%d')
value = value.toPandas().astype({'value':'float32'}).values

print(date)
```

- RLS 필터를 적용하여 데이터 이상치 탐색

```python
# feature들을 담을 크기. window size
windows = 15 # 초기에는 20으로 했었음
rls_filter = RLSFilter(windows=windows, lambda_=0.99, delta=1e-6, first_value = value[0])
# 이상치 데이터를 판단하는 리스트
anomaly_scores = np.zeros_like(value)
# y_predict = 예측값. 초기값은 value의 초기값으로.
y_predict = value[0]
print('전체 데이터 :', len(anomaly_scores))
# RLS 필터를 적용하여 이상치 탐색
for i in range(len(value)):
    # 예측값
    y_predict = rls_filter.update(value[i], y_predict)
    # 예측값들의 표준 편차.
    std = np.std(rls_filter.buffer)
    # 그래프를 그리기 위해, anomaly_scores에 이상치일 경우에는 1, 정상치일 경우에는 None을 넣음.
    # 예측값과 실제값의 차이가 표준 편차의 2.5배 이상의 차이가 난다면 이상치 1이라고 판단
    anomaly_scores[i] = 1 if np.abs(y_predict - value[i]) > 2.5 * std else None
# 이상치 행렬
an_scores = np.where(anomaly_scores == (1))[0]
print('이상치 행렬')
print(an_scores)
print('이상치 데이터 크기 : ', len(an_scores))
```

#### 시세와 레이드와 패키지간 연관성

- 던파의 꾸준한 패키지와 레이드 출시로 인해 시세가 안정적으로 유지 되고 있었음.

#### 결론
- 세라샵에서 판매하고 있는 '가브리엘의 계약' 제품을 참고하였을 때, 던파는 약 1:950 에서 1:1000 정도의 골드 비율을 원하고 있는 것 같음.
- 실제로 최근 며칠간 DVG를 계산했을때, 930골드 정도를 유지하였음.
- 지금까지처럼 패키지 출시와, 이벤트등을 유지 한다면, 던파의 재화가치는 크게 변동되지 않을 것으로 보여짐.

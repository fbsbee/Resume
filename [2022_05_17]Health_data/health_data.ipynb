### 이상치 정제 혹은 결측값 처리
* 혈당과 혈압은 매뉴얼에도 정상치가 나와있지 않기 때문에, 일단 데이터 보존.
* 그 외 혈색소, 혈청 등의 값들도 이상치가 존재하지만, 실제 나올 수 있는 값인지, 잘못 측정된 값인지 판단 할 수가 없어 데이터 보류.
* 청력, 음주, 흡연의 경우 categorical data 이기때문에 누락된 값들은 drop 처리
* 요단백이 가장 고민이었는데, drop시킬지 median값으로 처리할지 고민하다가, 누락된 데이터의 수가 1만건이 넘어가므로 median값으로 처리하였음.

* 시력은 9.9면 실명이므로, 이상치라고 판단하지 않음. 대신 시력이 정상치 값인 0.1~2.5사이 이외의 값인 0.0으로 치환
* 좌, 우 시력이 각각 따로 실명인 경우가 있을 수 있으므로. 각각 치환

* 시력의 경우 NaN 데이터를 boxplot의 median 값인 0.9로 대체
* data.loc[ data['시력(좌)'].isna() | data['시력(우)'].isna(), ['시력(우)', '시력(좌)'] ] = 0.9
* 허리 둘레의 경우 680 과 999인 경우가 존재하여 잘못 측정된 경우라고 판단, 두 경우 모두 성별코드 2에 연령대 코드가 9이므로, 성별코드가 2이고, 연령대 코드가 9인 사람들의 median 값으로 대체

* 허리 둘레의 값이 NaN인 경우가 100건 미만이기 때문에 drop 처리. 성별 코드와 연령대, 신장, 체중에 맞게 허리 둘레를 결측하기에 표본 데이터가 너무 적을 것이라 판단.


```python
# 이상치 값 정제 혹은 결측값 처리 1.
# 혈당과 혈압은 매뉴얼에도 정상치가 나와있지 않기 때문에, 1, 950과 같은 이상치인 경우에도 데이터 보존.
# 그 외 혈색소, 혈청 등의 값들도 이상치가 존재하지만, 실제 나올 수 있는 값인지, 잘못 측정된 값인지 판단 할 수가 없어 데이터 보류.
# 청력, 음주, 흡연의 경우 true or false 이기때문에 누락된 값들은 drop 처리
data = data.dropna(subset=['청력(좌)', '청력(우)', '음주여부', '흡연상태',])

# NaN의 숫자가 1% 내외라 drop을 할지 채워넣을지 고민.
# 요단백, 혈압, 혈당, 혈청 등과 같은 세부적인 수치는 나이와 성별에 따른 median 값으로 결측값 채워넣음
temp = ['요단백', '수축기 혈압', '이완기 혈압', '식전혈당(공복혈당)',
'혈색소', '혈청크레아티닌', '(혈청지오티)AST','(혈청지오티)ALT',
'감마 지티피', '시력(좌)', '시력(우)']
for feature in temp:
    data[feature].fillna(data.groupby(['연령대 코드(5세단위)', '성별코드'])[feature].transform('median'), inplace=True)
# 시력은 9.9면 실명이므로, 이상치라고 판단하지 않음. 대신 시력이 정상치 값인 0.1~2.5사이 이외의 값인 0.0으로 치환
# 좌, 우 시력이 각각 따로 실명인 경우가 있을 수 있으므로. 각각 치환
data.loc[ data['시력(좌)']==9.9, ['시력(좌)'] ] = 0.0
data.loc[ data['시력(우)']==9.9, ['시력(우)'] ] = 0.0
# 시력의 경우 NaN 데이터를 boxplot의 median 값인 0.9로 대체
# 허리 둘레의 경우 680 과 999인 경우가 존재하여 잘못 측정된 경우라고 판단, 두 경우 모두 성별코드 2에 연령대 코드가 9이므로, 성별코드가 2이고, 연령대 코드가 9인 사람들의 median 값으로 대체
data.loc[data['허리둘레']>150, ['허리둘레']] = data[(data['성별코드']==2) & (data['연령대 코드(5세단위)'] == 9)]['허리둘레'].median()
# 위 전처리 후 허리 둘레의 값이 NaN인 경우가 100건 미만이기 때문에 drop 처리. 성별 코드와 연령대, 신장, 체중에 맞게 허리 둘레를 결측하기에 표본 데이터가 너무 적을 것이라 판단.
data = data.dropna(subset=['허리둘레'])
```

### 학습모델 생성 및 model fit, evaluate 시행

* 기존 모델의 성능이 낮게 나와 트랜스포머 모델 채용해보았음


```python
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Normalization and Attention
    x = layers.LayerNormalization(epsilon=1e-6)(inputs)
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(x, x)
    x = layers.Dropout(dropout)(x)
    res = x + inputs

    # Feed Forward Part
    x = layers.LayerNormalization(epsilon=1e-6)(res)
    x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    return x + res

def build_model(
    input_shape,
    head_size,
    num_heads,
    ff_dim,
    num_transformer_blocks,
    mlp_units,
    n_classes=10,
    dropout=0.25,
    mlp_dropout=0.4,
):
    inputs = keras.Input(shape=input_shape)
    x = inputs
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    x = layers.GlobalAveragePooling1D(data_format="channels_first")(x)
    for dim in mlp_units:
        x = layers.Dense(dim, activation="relu")(x)
        x = layers.Dropout(mlp_dropout)(x)
    outputs = layers.Dense(n_classes, activation="softmax")(x)
    return keras.Model(inputs, outputs)
```

```python
input_shape = x_train.shape[1:]

model = build_model(
    input_shape,
    head_size=256,
    num_heads=4,
    ff_dim=4,
    num_transformer_blocks=4,
    mlp_units=[128],
    mlp_dropout=0.4,
    dropout=0.25,
    n_classes=data_y.shape[1]
)

model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    metrics=["sparse_categorical_accuracy"],
)
model.summary()
```

```python
# 모델 학습
# callbacks = [keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)]
check_point = './'
model_name = '{epoch:02d}-{val_sparse_categorical_accuracy:.4f}'
es = EarlyStopping(monitor='val_sparse_categorical_accuracy', patience=10, restore_best_weights=True)
mc = ModelCheckpoint(model_name, monitor='val_sparse_categorical_accuracy', verbose=1, save_best_only=True, mode='auto')
callbacks = [es, mc]
# callbacks = tf.keras.callbacks.EarlyStopping(monitor= 'val_loss', patience=200 * 0.1, verbose=2, restore_best_weights=True)
# es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=3)
# mc = ModelCheckpoint('best_model.h5', monitor = 'val_acc', mode='max', verbose=1, save_best_only=True)
history = model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=300,
    batch_size=128,
    callbacks=callbacks,
)
```

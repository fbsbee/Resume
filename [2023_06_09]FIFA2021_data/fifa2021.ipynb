## Data Cleaning & Transformation

- Convert the height and weight columns to numerical forms
- Remove the unnecessary newline characters from all columns that have them.
- Based on the 'Joined' column, check which players have been playing at a club for more than 10 years!
- 'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc.
- Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical
- Which players are highly valuable but still underpaid (on low wages)? (hint: scatter plot between wage and value)

```python
import pandas as pd
import numpy as np
import os
```

```python
folder = 'C:/Users/su/OneDrive - 청주대학교 (1)/datasets/before/2023'
file_name = '[2023_05_21]FIFA_21_players_dataset'
fifa1 = 'fifa21_raw_data.csv'
fifa2 = 'fifa21 raw data v2.csv'
```

```python
# fifa_data_one = pd.read_csv(os.path.join(folder, file_name, fifa1))
fifa_data = pd.read_csv(os.path.join(folder, file_name, fifa2))
```

```python
# drop_columns = ['photoUrl', 'playerUrl', 'Loan Date End', 'Release Clause']
drop_columns = ['photoUrl', 'playerUrl', 'Loan Date End']
fifa_data = fifa_data.drop(columns=drop_columns)
# fifa_data.drop(fifa_data.columns[[3,4,18]],axis = 1,inplace = True)
```

- 일부 열에 있는 '별' 문자 없애고 열을 숫자로 만듬
- 불필요한 줄바꿈 문자를 포함하는 모든 열에서 제거

```python
fifa_data.loc[:, 'Club']= fifa_data['Club'].str.replace('\n\n\n\n', '')
fifa_data.loc[:, 'IR']= fifa_data['IR'].str.replace('★','')
```

```python
fifa_data = fifa_data.rename(columns = {'↓OVA': 'OVA'})
```

```python
fifa_data.head()
```

- 높이 및 무게 열을 숫자 형식으로 변환

```python
print(fifa_data['Weight'].head())
print(fifa_data['Weight'].unique())
```

```python
fifa_data.loc[:, 'Weight'] = fifa_data['Weight'].str.replace('kg|lbs', '', regex=True)
# fifa_data.loc[:, 'Weight'] = fifa_data['Weight'].str.replace('lbs', '')
# fifa_data = fifa_data.astype({'Weight':'int'})
fifa_data['Weight'] = fifa_data['Weight'].astype(int)
```

```python
print(fifa_data['Weight'].head())
print(fifa_data['Weight'].unique())
```

```python
print(fifa_data['Height'].head())
print(fifa_data['Height'].unique())
```

```python
temp = fifa_data['Height'].str.contains("'")
fifa_data.loc[:, 'Height'] = fifa_data['Height'].str.replace('cm|"', '', regex=True)
# fifa_data.loc[:,'Height'] = fifa_data['Height'].str.replace('"', '')
for i in fifa_data[temp]['Height'].index:
    x, y = fifa_data['Height'][i].split("'")
    # feet : 30.48
    # inch : 2.54
    # fifa_data.loc[:, 'Height'][i] = round(int(x)*30.48 + int(y)*2.54)
    fifa_data.loc[i, 'Height'] = round(int(x)*30.48 + int(y)*2.54)
fifa_data['Height'] = fifa_data['Height'].astype(int)
```

```python
# fifa_data[fifa_data['Height'].str.contains('"')]['Height']
# fifa_data[temp]['Height'].head()
print(fifa_data['Height'].head())
print(fifa_data['Height'].unique())
```

```python
print(fifa_data['Hits'].head())
print(fifa_data['Hits'].unique())
```

```python
fifa_data.isnull().sum()
```

```python
fifa_data.loc[:, 'Hits'] = fifa_data['Hits'].fillna('0') # int 0으로 nan을 채우면 조작할때 nan형식으로 나올 수 있음. 그리고 어차피 str로 형변환해야함
fifa_data.loc[:, 'Hits'] = fifa_data['Hits'].astype(str) # 이미 float 형식으로 들어간 데이터가 존재 하므로 형변환.

temp = fifa_data['Hits'].str.contains("K")
# fifa_data['Hits'] = fifa_data['Hits'].str.replace('K', '')
fifa_data.loc[:, 'Hits'] = fifa_data['Hits'].str.replace('K', '')

fifa_data['Hits'] = fifa_data['Hits'].astype(float)
fifa_data.loc[temp, 'Hits'] = fifa_data[temp]['Hits']*1000
fifa_data['Hits'] = fifa_data['Hits'].astype(int)
```

```python
m = fifa_data['Release Clause'].str.contains("M")
k = fifa_data['Release Clause'].str.contains("K")
fifa_data['Release Clause'] = fifa_data['Release Clause'].str.replace('€|M|K', '', regex=True)
# fifa_data['Release Clause'] = fifa_data['Release Clause'].str.replace('M', '')
# fifa_data['Release Clause'] = fifa_data['Release Clause'].str.replace('K', '')
fifa_data['Release Clause'] = fifa_data['Release Clause'].astype(float)

fifa_data.loc[m, 'Release Clause'] = fifa_data[m]['Release Clause']*1000000
fifa_data.loc[k, 'Release Clause'] = fifa_data[k]['Release Clause']*1000

fifa_data['Release Clause'] = fifa_data['Release Clause'].astype(int)
```

```python
m = fifa_data['Value'].str.contains("M")
k = fifa_data['Value'].str.contains("K")
fifa_data['Value'] = fifa_data['Value'].str.replace('€|M|K', '', regex=True)
fifa_data['Value'] = fifa_data['Value'].astype(float)

fifa_data.loc[m, 'Value'] = fifa_data[m]['Value']*1000000
fifa_data.loc[k, 'Value'] = fifa_data[k]['Value']*1000
fifa_data['Value'] = fifa_data['Value'].astype(int)
```

```python
print(fifa_data['Value'].head())
print(fifa_data['Value'].unique())
```

```python
print(fifa_data['Wage'].head())
print(fifa_data['Wage'].unique())
```

```python
k = fifa_data['Wage'].str.contains("K")
fifa_data['Wage'] = fifa_data['Wage'].str.replace('€|K', '', regex=True)
fifa_data['Wage'] = fifa_data['Wage'].astype(int)

fifa_data.loc[k, 'Wage'] = fifa_data[k]['Wage']*1000
# fifa_data['Wage'] = fifa_data['Wage'].astype(int)
```

```python
print(fifa_data['Wage'].head())
print(fifa_data['Wage'].unique())
```

- 어떤 선수들이 높은 가치를 가지고 있지만 여전히 낮은 임금을 받고 있는지 확인( Value와 Wage )

```python
import plotly.express as px
import plotly.graph_objects as go

fifa_data2 = fifa_data.copy()
# 데이터 정규화
fifa_data2['Value'] = (fifa_data2['Value'] - fifa_data2['Value'].min())/(fifa_data2['Value'].max() - fifa_data2['Value'].min())
fifa_data2['Wage'] = (fifa_data2['Wage'] - fifa_data2['Wage'].min())/(fifa_data2['Wage'].max() - fifa_data2['Wage'].min())
f1 = fifa_data2[fifa_data2['Value']<=0.15]
f2 = fifa_data2[fifa_data2['Value']>0.15]
fig = px.scatter(f1, x='Value', y='Wage', color='Wage', color_continuous_scale = px.colors.sequential.Sunsetdark)
fig2 = px.scatter(f2, x='Value', y='Wage', color='Wage', color_continuous_scale = px.colors.sequential.Sunsetdark)

fig.show()
fig2.show()
```

```python
ch = (fifa_data2['Wage']==0) & (fifa_data2['Value']==0)
result = fifa_data2[~ch]
result = result[(result['Wage']/result['Value'])<0.5] # 0.5 미만을 저임금이라고 가정하였음.
```

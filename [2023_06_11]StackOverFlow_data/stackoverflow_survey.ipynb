- 직업 별 평균 연봉

```python
pub_real.DevType.unique()
```

```python
# devtype 간소화. for문으로 전체 순회.
for i in pub_real.index:
    devtype = re.split('[,; ]', pub_real.loc[i, 'DevType'])[0]
    # print(devtype)
    pub_real.loc[i, 'DevType'] = devtype
print('convert done')
```

```python
# 직업 별 평균 연봉
result = pub_real.groupby('DevType').mean()
result = result.sort_values(by='ConvertedCompYearly')
# 인덱스에 있는 devtype 다시 column으로.
result.reset_index(level=['DevType'], inplace = True)
result[['DevType', 'ConvertedCompYearly']]
```

```python
fig = px.bar(result, x='DevType', y='ConvertedCompYearly', )
fig.show()
```

#### 직군별, 원격, 금액, mental 데이터의 평균.(상관관계)
- remote work data

```python
pub_real.RemoteWork.unique()
```

```python
pub_real.RemoteWork.isna().sum()
```

```python
pub_real = pub_real.dropna(axis=0, subset=['RemoteWork'])
```

```python
# 원격을 많이 할 수록 점수(수치)를 더 높게 부여.
def get_remote_int(row):
    if isinstance(row.RemoteWork, str):
        if 'Hybrid' in row.RemoteWork:
            return 1
        elif 'remote' in row.RemoteWork:
            return 2
        else:
            return 0
    else:
        return 0
rw = pub_real.apply(get_remote_int, axis=1)
pub_real.RemoteWork = rw
```

- mental health data

```python
pub_real.MentalHealth.unique()
```

```python
# None of the above , nan 은 0 나머지는 1로 변환.
# pub_real.loc[:, 'MentalHealth'] = pub_real['MentalHealth'].fillna('0')
# pub_real.loc[:, 'MentalHealth'] = pub_real['MentalHealth'].str.replace('None of the above', '')
# pub_real['MentalHealth'] = pub_real['MentalHealth'].astype(int)

# 해당이 없으면 상태가 좋은걸로 판단.
def get_health_bool(row):
    if isinstance(row.MentalHealth, str):
        return 0 if "None of the above" in row.MentalHealth else 1
    else:
        return 0
mh = pub_real.apply(get_health_bool, axis=1)
pub_real['MentalHealth'] = mh
```

```python
pub_real[['DevType', 'ConvertedCompYearly', 'MentalHealth', 'RemoteWork']].head()
```

```python
# 직군별, 원격별, 금액별, mental 수치의 평균
result = pub_real.groupby('DevType')
result = result[['ConvertedCompYearly', 'MentalHealth', 'RemoteWork']].agg(['mean'])
# 인덱스에 있는 devtype 다시 column으로.
result.reset_index(level=['DevType'], inplace = True)
# 'mean' 컬럼 drop 시킴.
result = result.droplevel(axis=1, level=1, )
result = result.sort_values(by='ConvertedCompYearly')
result
```

```python
result['ConvertedCompYearly'] = (result['ConvertedCompYearly'] - result['ConvertedCompYearly'].min())/(result['ConvertedCompYearly'].max() - result['ConvertedCompYearly'].min())
result['RemoteWork'] = (result['RemoteWork'] - result['RemoteWork'].min())/(result['RemoteWork'].max() - result['RemoteWork'].min())
result
```

```python
result.corr(method='pearson')
```

```python
fig = px.imshow(result.corr(method='pearson'), text_auto=True, aspect="auto")
fig.show()
```

- 원격 작업과 연봉, 원격 작업과 정신 건강은 상관관계가 높은 편이지만, 연봉과 정신건강은 관계가 매우 낮음을 보여주었음.
- 연봉이 높은 곳일수록 원격 작업을 많이 한다고 생각할 수 있으며, 원격 작업을 많이 할수록 정신 건강도 좋은 편이란 것을 알 수 있었음.
